```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import deepchem as dc
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
plt.rcParams['figure.dpi'] = 300
```

```python
gabaa_agonist = pd.read_csv('../data/refined_gabaa.csv')[0:827]
```

```python
gabaa_nonagonist = pd.read_csv('../data/refined_gabaa.csv')[827:]
```

# refined_gabaa_ECFP

```python
featurizer_ecfp = dc.feat.CircularFingerprint(size=1024,radius=4)
X1 = featurizer_ecfp.featurize(gabaa_agonist['smiles'])
X2 = featurizer_ecfp.featurize(gabaa_nonagonist['smiles'])
```

```python
data_ecfp = np.concatenate((X1, X2), axis=0)

# 运行t-SNE算法
tsne = TSNE(n_components=2)
transformed_data = tsne.fit_transform(data_ecfp)

# 绘制结果
sns.scatterplot(x=transformed_data[:, 0], y=transformed_data[:, 1], hue=['agonist']*827 + ['non-agonist']*827)
```

# refined_gabaa_RDkit

```python
featurizer_rdkit = dc.feat.RDKitDescriptors()
X1 = featurizer_rdkit.featurize(gabaa_agonist['smiles'])
X2 = featurizer_rdkit.featurize(gabaa_nonagonist['smiles'])
```

```python
data_rdkit = np.concatenate((X1, X2), axis=0)

# 运行t-SNE算法
tsne = TSNE(n_components=2)
transformed_data = tsne.fit_transform(data_rdkit)

# 绘制结果
sns.scatterplot(x=transformed_data[:, 0], y=transformed_data[:, 1], hue=['agonist']*827 + ['non-agonist']*827)
```
