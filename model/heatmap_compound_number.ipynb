```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data = {'Alcohols_ET': [6, 6, 5, 6, 7, 4, 5],
        'Alcohols_PT': [6, 8, 7, 8, 11, 9, 7],
        'Aldehydes_ET': [9, 8, 6, 7, 9, 8, 7],
        'Aldehydes_PT': [9, 9, 10, 12, 11, 11, 11],
        'Acids_ET': [2, 2, 3, 3, 6, 3, 4],
        'Acids_PT': [3, 3, 4, 4, 5, 5, 4],
        'Esters_ET': [0, 3, 4, 2, 3, 4 ,5],
        'Esters_PT': [3 ,3 ,4 ,2 ,1 ,4 ,3],
        'Ketones ET': [2, 3, 2, 2, 2, 2, 2],
        'Ketones PT': [5, 3, 4, 7, 3, 7, 4],
        'Phenols ET': [1, 1, 0, 1, 1, 1, 1],
        'Phenols PT': [3, 2, 2, 2, 1, 2, 3],
        'Hydrocarbons ET': [2, 1, 0 ,1 ,2 ,2 ,2],
        'Hydrocarbons PT': [7 ,7 ,6 ,7 ,11 ,8 ,10],
        'Heterocyclic rings ET': [1 ,1 ,2 ,2 ,0 ,2 ,1],
        'Heterocyclic rings PT': [2 ,1 ,1 ,4 ,1 ,2 ,3]
       }
index = ['FL', 'ES', 'TB', 'MO', 'GK', 'KP', 'KL']

df = pd.DataFrame(data=data,index=index)

```

```python
fig1 = plt.figure(figsize=(16,5))
sns.heatmap(df,cmap='YlGnBu',annot = True)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tick_params(axis='both', which='both', length=0)
ax = plt.gca()
ax.set_xticklabels(ax.get_xticklabels(), ha='right')
plt.show()
```

```python
fig1.savefig('heatmap.png', dpi=300,bbox_inches = 'tight')
```
